```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('banking.csv')
```

```python
df
```

```python
df.head()
```

```python
bins =[0,100000,300000,float('inf')]
labels =['Low','Med','High']
df['Income Band'] =pd.cut (df['Estimated Income'],bins=bins,labels=labels,right=False)
```

```python
df['Income Band'].value_counts()
```

```python
plt.figure()
plt.hist(df['Age'].sample(50), bins=20)
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
top_occ = df["Occupation"].value_counts().head(10)


plt.figure()
top_occ.plot(kind="bar")
plt.title("TOP 10 Occupations")
plt.xlabel("Occupation")
plt.ylabel("Number of Customer")
plt.show()
```

```python
avg_income = df.groupby("Loyalty Classification")["Estimated Income"].mean()

plt.figure()
avg_income.plot(kind="bar")
plt.title("Average Income by Loyalty Classification")
plt.ylabel("Average Income")
plt.show()
```

```python
plt.figure()
plt.hist(df["Credit Card Balance"], bins=20)
plt.title("Credit Card Balance Distribution")
plt.xlabel("Balance")
plt.ylabel("Frequency")
plt.show()
```
