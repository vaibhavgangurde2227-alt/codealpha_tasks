```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('banking.csv')
```

```python
df
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df , x='BRId',hue = 'GenderId')
plt.title('BRID VS GenderId')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df ,x='IAId',hue='GenderId')
plt.title('IAId vs GenderId')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Amount of Credit Cards', hue='GenderId')
plt.title('Credit Cards vs GenderId')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Occupation', hue='GenderId')
plt.title('Occupation vs GenderId')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Fee Structure', hue='GenderId')
plt.title('Fee Structure vs GenderId')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Loyalty Classification', hue='GenderId')
plt.title('Loyalty Classification vs GenderId')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

categorical_cols = [
    'BRId',
    'GenderId',
    'IAId',
    'Amount of Credit Cards',
    'Nationality',
    'Occupation',
    'Fee Structure',
    'Loyalty Classification'
]

for i, predictor in enumerate(categorical_cols):
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=predictor, hue='Nationality')
    plt.title(f'{predictor} vs GenderId')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings']

plt.figure(figsize=(8, 4))

for i, col in enumerate(numerical_cols):
    plt.subplot(1, 2, i+1)   # 1 row, 2 columns
    sns.histplot(df[col], kde=True)
    plt.title(col)

plt.tight_layout()
plt.show()
```
